```python
# modules we'll use
import pandas as pd
import numpy as np
import seaborn as sns
import datetime

# read in our data
landslides = pd.read_csv("../input/landslide-events/catalog.csv")

# set seed for reproducibility
np.random.seed(0)
```

```python
landslides.head()
```

```python
# print the first few rows of the date column
print(landslides['date'].head())
```

```python
# check the data type of our date column
landslides['date'].dtype
```

"O" is the code for "object"

```python
# create a new column, date_parsed, with the parsed dates
landslides['date_parsed'] = pd.to_datetime(landslides['date'], format="%m/%d/%y")
```

```python
# print the first few rows
landslides['date_parsed'].head()
```

We use the following in case there was different data entreies (multiple date formats)
landslides['date_parsed'] = pd.to_datetime(landslides['Date'], infer_datetime_format=True)

We can extract part of the date if we like to ... 

```python
# get the day of the month from the date_parsed column
day_of_month_landslides = landslides['date_parsed'].dt.day
day_of_month_landslides.head()
```

If we tried to get the same information from the original "date" column, we would get an error: AttributeError: Can only use .dt accessor with datetimelike values. This is because dt.day doesn't know how to deal with a column with the dtype "object". Even though our dataframe has dates in it, we have to parse them before we can interact with them in a useful way.

Now let think to visualize the data to detect if there is any error. For instance, Reading the date date through the previous
steps may come up with a common error which is mixing between the day and the month. 

months are from 1 to 12
days are between 1 to 28, 30, 31 in general.

If we presented days via a histogram then we should see values byond 1-12 as there is no reasone to let us think 
that the values are related to the first 12 days of each month. 

```python
# remove na's
day_of_month_landslides = day_of_month_landslides.dropna()

# plot the day of the month
sns.distplot(day_of_month_landslides, kde=False, bins=31)
```

The histogram shows that the code does not mix between days and months
